```python
import pandas as pd
```

```python
df7 = pd.read_csv('./Olist_sales_dataset/df7_olist_products_dataset.csv')
```

```python
# View the first or last few rows to understand the structure.

df7.head()
```

```python
# Check data types and see if there are any missing values.

df7.info()
```

## Handling Duplicates:

```python
df7.loc[:, 'product_id'].drop_duplicates(inplace=True)
```

## Handling Missing Data

```python
#  Check for missing values in each column

df7.isnull().sum()
```

```python
# Remove rows with missing values 

clean_df7 = df7.dropna()
clean_df7
```

```python
clean_df7.isnull().sum()
```

## Column Renaming and Reordering:

```python
# Rename columns for clarity and reorder them as needed:

clean_df7.rename(columns={'product_id':'Product id',
                          'product_category_name':'Product category',
                         'product_name_lenght':'Product name lenght',
                           'product_description_lenght':'Product description lenght',
                          'product_photos_qty':'Product photos qty',
                         'product_weight_g':'Product weight g',
                         'product_length_cm': 'Product length cm',
                         'product_height_cm': 'Product height cm',
                         'product_width_cm': 'Product width cm'}, inplace=True)

clean_df7
```

```python
clean_df7.to_csv('cleaned_dataset_df7.csv', index=False)
```
