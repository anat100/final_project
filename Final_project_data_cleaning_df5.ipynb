## Handling Duplicates:

```python
df5.loc[:, 'review_id'].drop_duplicates(inplace=True)
```

```python
df5.loc[:, 'order_id'].drop_duplicates(inplace=True)
```

## Handling Missing Data

```python
#  Check for missing values in each column

df5.isnull().sum()
```

```python
# Remove rows with missing values 

clean_df5 = df5.dropna()
clean_df5
```

```python
clean_df5.isnull().sum()
```

## Column Renaming and Reordering:

```python
# Rename columns for clarity and reorder them as needed:

clean_df5.rename(columns={'review_id':'Review id',
                          'order_id':'Order id',
                         'review_score':'Review score',
                           'review_comment_title':'Review Comment title',
                         'review_comment_message':'Review Comment message'}, inplace=True)

clean_df5
```
